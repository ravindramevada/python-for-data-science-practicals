# **Practical - 01**

Write a program to create a number converter, that can convert numbers between binary, decimal, octal and hexadecimal. 

```
base_inp = int(input('Please enter the base of your input number: '))
my_dict = {2: 'Binary', 8: 'Octal', 10: 'Decimal', 16: 'Hexadecimal'}
num = input('Please enter the {} number: ' .format(my_dict[base_inp]))
tmp = int(num, base_inp)
base_out = int(input('Please enter the base of your output number: '))
if base_out == 2:
  print(num, 'in binary is: ', bin(tmp))
elif base_out == 8:
  print(num, 'in octal is: ', oct(tmp))
elif base_out == 10:
  print(num, 'in decimal is: ', tmp)
elif base_out == 16:
  print(num, 'in hexadecimal is: ', hex(tmp))
```

# **Practical - 03**

Write a program to solve a quadratic equation.

```
from math import sqrt

a = int(input('Enter a: '))
b = int(input('Enter b: '))
c = int(input('Enter c: '))

discriminant = (b * b) - (4 * a * c)
print('Discriminant = ', discriminant)
if discriminant > 0:
  root1 = ((-b + sqrt(discriminant)) / (2 * a))
  root2 = ((-b - sqrt(discriminant)) / (2 * a))
  print('Two distinct real roots exists: root1 = %.2f and root2 = %.2f' %(root1, root2))
elif discriminant == 0:
  root1 = root2 = -b / (2 * a)
  print('Two real and equal roots exists: root1 = %.2f and root2 = %.2f' %(root1, root2))
elif discriminant < 0:
  root1 = root2 = -b / (2 * a)
  imaginary = sqrt(-discriminant) / (2 * a)
  print('Two distinct complex roots exists: root1 = %.2f+i%.2f and root2 = %.2f-i%.2f' %(root1, imaginary, root2, imaginary))
```

# **Practical - 05**

Write a program to count the frequency of bigrams and four grams from a given string.

```
def cnt_freq(my_str, n):
  ngrams = []
  for i in range(len(my_str)-(n-1)):
    ngrams.append(my_str[i:i+n])
  word_freq = []
  for j in ngrams:
    word_freq.append(ngrams.count(j))
  return list(zip(ngrams, word_freq)) 

my_str = 'Ravindra Ravindra'
bigrams = 2
four_grams = 4
print('Bigrams: ', cnt_freq(my_str, bigrams))
print('Four grams: ', cnt_freq(my_str, four_grams))
```

# **Practical - 10**

Create function word_occur which takes a string as an argument and returns a dictionary where keys are words of the string and values are the occurrence of each word in the string. 

```
def word_occur(my_str):
  my_dict = {}
  my_str = my_str.lower().split()
  for word in my_str:
    if word in my_dict:
      my_dict[word] += 1
    else:
      my_dict[word] = 1
  return my_dict

my_str = 'My Name is Ravindra Rav Indra Ravindra'
print(word_occur(my_str))
```

# **Practical - 11**


Create a class called Shape, which has three subclasses say Triangle, Rectangle, Circle. Define one method area()in the shape class and override this area() in these three subclasses to calculate for specific objects i.e., area() of the Triangle subclass should calculate the area of a triangle, etc. Same for Rectangle and Circle.

```
from math import pi

class Shape():
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def area(self):
    pass
    
class Triangle(Shape):
  def __init__(self, base, height):
    self.base = base
    self.height = height
    Shape.__init__(self, base, height)
  
  def area(self):
    print('Area of the triangle is: ', 1/2*self.base*self.height)

class Rectangle(Shape):
  def __init__(self, length, breadth):
    self.length = length
    self.breadth = breadth
    Shape.__init__(self, length, breadth)
  
  def area(self):
    print('Area of the rectangle is: ', self.length*self.breadth)

class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius
    Shape.__init__(self, radius, radius)

  def area(self):
    print('Area of the circle is: ', pi*self.radius*self.radius)

t = Triangle(2, 3)
t.area()
r = Rectangle(2, 3)
r.area()
c = Circle(5)
c.area()
```

# **Practical - 17**

Find out mean, median and mode of list of values without using any function.

```
my_list = [31, 100, 36, 12, 69, 38, 36, 76, 61, 29]
mean = sum(my_list) / len(my_list)
median = sorted(my_list)[len(my_list) // 2]
mode = max(my_list, key=my_list.count)
print('Mean:', round(mean), '\nMedian:', median, '\nMode:', mode)
```

# **Practical - 18**

Find out range, variance and standard deviation of list of values without using any function.

```
my_list = [31, 100, 36, 12, 69, 38, 36, 76, 61, 29]
range = max(my_list) - min(my_list)
mean = sum(my_list) / len(my_list) 
variance = sum([((x - mean) ** 2) for x in my_list])/len(my_list) 
std_dev = variance ** 0.5
print('Range:', range, '\nVariance:', round(variance, 2), '\nStandard Deviation:', round(std_dev, 2))
```

# **Practical - 19**

Find out correlation between two lists of values using spearman correlation and pearson correlation.

```
from scipy import stats
list1 = [13, 16, 12, 21, 7]
list2 = [30, 29, 28, 11, 16]
correlation1, p_value1 = stats.spearmanr(list1, list2)
correlation2, p_value2 = stats.pearsonr(list1, list2)
print('Spearman Correlation:\ncorrelation', correlation1, 'p-value:', p_value1)
print('Pearson Correlation:\ncorrelation', correlation2, 'p-value:', p_value2)
```
